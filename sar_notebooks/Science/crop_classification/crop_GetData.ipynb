```python
import gdal
```

```python
#!/bin/bash

for f in *.grd do
    echo $f
    INPUT=$f
    OUTPUT=${f}_clip.tif
    gdal_translate $INPUT $OUTPUT -projwin -121.5198 37.9766 -121.3487 37.8420
done
```
